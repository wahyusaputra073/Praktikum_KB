```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Car details v3.csv")
df = df.iloc[:3000]
```

------------------------------------------------------------------------------------------------------------------------------

### Dataset ini berisi informasi tentang mobil bekas yang terdaftar di berbagai situs web

```python
df.head(10)
```

------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(10,5))
x = list(df['fuel'])
y = list(df['km_driven'])
plt.bar(x,y,color="blue")

plt.title("Perbandingan jarak tempuh pada setiap jenis bahan bakar")
plt.ylabel('Jarak (Km.)')
plt.xlabel('Jenis Bahan Bakar')
plt.show()
```

Grafik ini memperlihatkan bahan bakar paling banyak yang menempuh jarak terjauh. dan grafik ini menyimpulkan bahwa mobil bekas yang tersedia paling banyak menggunakan jenis bahan bakar Diesel untuk perjalanan jarak jauh

------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x= 'fuel',palette='flare')
plt.title("Perbandingan jumlah mobil bekas pada setiap jenis bahan bakar")
plt.xlabel('Jenis Bahan Bakar')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik ini memperlihatkan jumlah mobil bekas yang menggunakan jenis bahan bakar tertentu. Kesimpulan dari ini adalah bahan bakar Diesel yang paling banyak digunakan

------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x= 'transmission',palette='flare')
plt.title("Perbandingan jumlah mobil bekas menurut jenis transmisi")
plt.xlabel('Jenis Transmission')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik ini memperlihatkan jumlah mobil bekas yang menggunakan jenis transmisi tertentu. Kesimpulan dari grafik ini adalah transmisi manual yang paling banyak dipakai pada mobil bekas

------------------------------------------------------------------------------------------------------------------------------

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df,x= 'transmission',palette='flare',hue='fuel')
plt.title("Perbandingan jenis bahan bakar pada setiap transmisi")
plt.xlabel('Jenis Transmisi')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik ini memperlihatkan jumlah jenis bahan bakar yang dipakai pada setiap jenis transmisi. Kesimpulan dari grafik ini adalah bahan bakar jenis Diesel paling banyak dipakai pada jenis transmisi manual

------------------------------------------------------------------------------------------------------------------------------

```python
pieData = df["fuel"].value_counts().rename_axis("fuel").reset_index(name="id_count")
```

```python
plt.figure(figsize=(11,11))
plt.pie(pieData["id_count"],labels=pieData["fuel"],startangle=120,autopct='%.1f%%')
plt.xlabel('Jenis Bahan Bakar')
plt.ylabel('Kuantitas')
plt.show()
```

Grafik di bawah ini memperlihatkan jumlah jenis bahan bakar yang dipakai pada setiap jenis transmisi. Kesimpulan dari grafik ini adalah bahan bakar jenis Diesel paling banyak dipakai pada jenis transmisi manual diiuti dengan bahan bakar jenis Petrol yang mencapai 46.8%
